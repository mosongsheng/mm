```python
import pandas as pd  
import seaborn as sns  
from scipy.stats import chi2_contingency  
  
# 读取SPSS数据文件  
file_path_spss = '国家认同数据清理后.sav'  
df_spss = pd.read_spss(file_path_spss)  
  
# 为了演示，假设'年级'是我们要选定的年级，'政治面貌'是我们要分析的政治面貌变量，  
# '某个类别变量'是我们要分析的其他变量  
selected_grade = '年级'  
selected_variable = '会打多少分'  
  
# 创建交互分类表  
cross_tab = pd.crosstab(index=[df_spss[selected_grade], df_spss['政治面貌']],   
                        columns=[df_spss[selected_variable]],   
                        margins=True, normalize=True)  # All 行和列的总和  
print(cross_tab)  
  
# 进行卡方检验  
chi2, p, dof, expected = chi2_contingency(cross_tab)  
print('卡方检验结果：')  
print('卡方统计量:', chi2)  
print('p值:', p)  
```
