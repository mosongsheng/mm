```python
import matplotlib.pyplot as plt  
  
# 定义数据  
labels = ['Apple', 'Banana', 'Cherry', 'Dates']  
sizes = [15, 30, 45, 10]  
  
# 使用def定义一个函数来创建饼图  
def create_pie_chart(labels, sizes):  
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')  
    plt.show()  
  
# 调用函数创建饼图  
create_pie_chart(labels, sizes)
```
